### Plot Average Runtimes vs Thread Amount

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("resultMetrics.csv")

plt.figure(figsize=(6,4))
plt.plot(data["Threads"], data["Mean Runtimes"], marker='.')
plt.title('Bridge Simulation Performance Scaling: Runtime vs. Threads', fontsize=14)
plt.ylabel('Average Runtime (seconds)', fontsize=12)
plt.xlabel('Thread Amount', fontsize=12)
plt.tight_layout()
plt.show()
```

### Alternative: Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("resultMetrics.csv")

plt.figure(figsize=(6,4))
plt.bar(data["Threads"], data["Mean Runtimes"])
plt.title('Bridge Simulation Performance Scaling: Runtime vs. Threads', fontsize=14)
plt.ylabel('Average Runtime (seconds)', fontsize=12)
plt.xlabel('Thread Amount', fontsize=12)
plt.tight_layout()
plt.show()
```

### Plot Observed & Ideal Speedups vs Thread Amount

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("resultMetrics.csv")

plt.figure(figsize=(6, 4))
plt.plot(data["Threads"], data["Observed Speedup"], marker='.', label="Observed")
plt.plot(np.linspace(0, 11), np.linspace(0, 11), color="green", linestyle="--", label="Ideal")
plt.title('Bridge Simulation Performance Scaling: Speedup vs. Threads', fontsize=14)
plt.ylabel('Runtime Speedup Ratio', fontsize=12)
plt.xlabel('Thread Amount', fontsize=12)
plt.tight_layout()
plt.legend()
plt.show()
```

### Plot Effenciency vs Threads

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("resultMetrics.csv")

plt.figure(figsize=(6,4))
plt.plot(data["Threads"].to_numpy()[1:], data["Efficiency"].to_numpy()[1:] * 100, marker='.')
plt.title('Bridge Simulation Performance Scaling: Efficiency vs. Threads', fontsize=14)
plt.ylabel('Parallel Efficiency (%)', fontsize=12)
plt.yticks(np.linspace(0, 100, 11))
plt.xlabel('Thread Amount', fontsize=12)
plt.tight_layout()
plt.show()
```
